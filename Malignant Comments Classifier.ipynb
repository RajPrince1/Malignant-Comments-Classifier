# Data Cleaning

Steps for "comment_text" columndf_train[i] = df_train[i].astype('category')

Remove HTML tags,Remove special characters, Convert everything to lowercase, Remove stopwords, Lemmatization

```python
#importing libraries as required
import re
import nltk
import string
from nltk.corpus import stopwords
```

# 1- Removing the HTML tags

```python
# Regex rule : "<.*?>"

def clean(text):
    cleaned = re.compile(r"<.*?>")
    return re.sub(cleaned," ",text)

train_df.comment_text = train_df.comment_text.apply(clean)
train_df.comment_text[1]
```

# 2- Removing special character like punctuation

```python
# We can use string.punctuation or manually

def is_special(text):
    empty = ''
    for i in text:
        if i.isalpha():
            empty = empty + i
        else:
            empty = empty + ' '
    return empty

train_df.comment_text = train_df.comment_text.apply(is_special)
train_df.comment_text[1]
```

# 3-Converting everything to lowercase

```python

def lower(text):
    return text.lower()

train_df.comment_text = train_df.comment_text.apply(lower)
train_df.comment_text[1]
```

# 4-Removing the stopwords

```python
import nltk
nltk.download('punkt')

from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize

def remove_stopwords(text):
    stop_words = set(stopwords.words('english') + ['u', 'ü', 'ur', '4', '2', 'im', 'dont', 'doin', 'ure','yo','Ja'])
    words = word_tokenize(text)
    return [w for w in words if w not in stop_words]

train_df.comment_text = train_df.comment_text.apply(remove_stopwords)
train_df.comment_text[1]
```

# 5-Lemmatization

```python
from nltk.stem import WordNetLemmatizer
nltk.download('wordnet')

def lema_text(text):
    lema = WordNetLemmatizer()
    return " ".join([lema.lemmatize(w) for w in text])

train_df.comment_text = train_df.comment_text.apply(lema_text)
train_df.comment_text[1]
```

```python
# Creating new column (cleaned_text) after removing puncuations,stopwords from news column to check how much data is cleaned
train_df['cleaned_text'] = train_df.comment_text.str.len()
train_df.head()
```

```python
# Total length removal
print("Origional Length:",train_df.comment_length.sum())
print("Cleaned Length:",train_df.cleaned_text.sum())
print("Total Words Removed:",(train_df.comment_length.sum()) - (train_df.cleaned_text.sum()))
```

# Converting the text into vectors using TF-IDF Vectorizer

```python
# Importing the library and converting it into vectors
from sklearn.feature_extraction.text import TfidfVectorizer
tf_vec = TfidfVectorizer(max_features = 15000, stop_words='english')

#Independent Features
X = tf_vec.fit_transform(train_df['comment_text'])

#Dependent Feature
y = train_df['label']
#Checking the distribution of the labels
print("Count of Benign and Malignant Comments:\n",train_df.label.value_counts())
```

# Model Training

```python
#checking the shape of input and target variable
print(X.shape,'\t\t',y.shape)
```

```python
#Splitting the data into training and testing data

from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(X,y,test_size=.25,random_state=50,stratify=y)
```

```python
#Checking the shape

print(x_train.shape,'\t\t',x_test.shape)
```

# Logistic Regression
Cross Validation like KFold and Hyperparameter tunning

```python
# Importing the model library
from sklearn.linear_model import LogisticRegression

#Importing the error metrics
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
from sklearn.model_selection import GridSearchCV, KFold

log_class=LogisticRegression()
grid={'C':10*np.arange(-2,3),'penalty':['l1','l2']}
cv=KFold(n_splits=15,random_state=None,shuffle=False)

clf=GridSearchCV(log_class,grid,cv=cv,n_jobs=-1,scoring='f1_macro')
clf.fit(x_train,y_train)
```

```python
y_pred_train = clf.predict(x_train)
print('Training accuracy is {}\n'.format(accuracy_score(y_train, y_pred_train)))
y_pred_test=clf.predict(x_test)
print('Test accuracy is {}\n'.format(accuracy_score(y_test,y_pred_test)))
print(confusion_matrix(y_test,y_pred_test))
print(classification_report(y_test,y_pred_test))
```

# Under Sampling

By using the under sampling we are trying to reduce the points of maximum labels.

```python
from collections import Counter
Counter(y_train)
```

```python
# Importing the Undersampling library
from imblearn.under_sampling import NearMiss
```

```python
ns = NearMiss(.8)
x_train_ns,y_train_ns = ns.fit_resample(x_train,y_train)

print("The number of classes before fit {}".format(Counter(y_train)))
print("The number of classes after fit {}".format(Counter(y_train_ns)))
```

# Using XGBoost with Undersampled data


```python
from xgboost import XGBClassifier
xgb_us = XGBClassifier()
xgb_us.fit(x_train_ns,y_train_ns)
```

```python
y_pred_train = xgb_us.predict(x_train_ns)
print('Training accuracy is {}\n'.format(accuracy_score(y_train_ns, y_pred_train)))
y_pred_test=xgb_us.predict(x_test)
print('Test accuracy is {}\n'.format(accuracy_score(y_test,y_pred_test)))
print(confusion_matrix(y_test,y_pred_test))
print(classification_report(y_test,y_pred_test))
```

From the above result we can clearly see that our under sampled model gives very poor result. So don't use under sampling unless, untill our dataset is very small.

# Over Samping


By using the over sampling we are trying to increase the points of minimum labels.

```python
# Importing the Oversampling library
from imblearn.over_sampling import RandomOverSampler
```

```python
os = RandomOverSampler(0.75)
x_train_os,y_train_os = os.fit_resample(x_train,y_train)
print("The number of classes before fit {}".format(Counter(y_train)))
print("The number of classes after fit {}".format(Counter(y_train_os)))
```
